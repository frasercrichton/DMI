```python
import pandas as pd
import os 
from pathlib import Path
import replicate
import pandas as pd

source_dir = '../data/source/'
processed_dir = '../data/processed/'

pd.set_option('display.max_colwidth', None)

os.makedirs(processed_dir + 'images', exist_ok = True)
```

```python
def get_prompt(image_location):
    return replicate.run(
        "methexis-inc/img2prompt:50adaf2d3ad20a6f911a8a9e3ccf777b263b8596fbd2c8fc26e8888f8a0edbb5",
        input={
            "image": open(image_location, "rb")
        }
    )
```

```python
dmi_images_folder = f"{processed_dir}images/dmi"

pathlist = Path(dmi_images_folder).glob('**/*.jpg')

def get_prompts_for_all_images_in_folder(folder_name):

    images_with_prompts = []

    for path in pathlist:
        prompt = get_prompt(dmi_images_folder + f"/{str(path.name)}")
        print(prompt)
        image_with_prompt = {"image_name": str(path.name), "prompt": prompt}

        images_with_prompts.append(image_with_prompt)

images_with_prompts = get_prompts_for_all_images_in_folder(dmi_images_folder)
print(images_with_prompts) 

df = pd.DataFrame(images_with_prompts)
df.to_csv(f"{processed_dir}/images_with_prompts.csv")
```

```python
def get_image_name(text):
    file_name = text.split('?')[0].split('/')[-1]
    print(file_name)   
    return file_name

images_with_prompts_df = pd.read_csv(f"{processed_dir}" + 'images_with_prompts.csv')

instagram_images_df = pd.read_csv(f"{source_dir}" + "AIphotography-top10k-instagram-230925 - top 200 posts.csv")
instagram_images_df['image_name'] = instagram_images_df['Photo'].apply(lambda row : get_image_name(text = row))

df = pd.merge(images_with_prompts_df, instagram_images_df, on=['image_name'])
df.to_csv(f"{processed_dir}" + 'AIphotography-200.csv')
```
